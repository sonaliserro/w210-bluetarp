```python
def replace_missing(shapefile):
    shapefile.RPL_THEMES.replace(-999.0, np.nan, inplace=True)
    shapefile.RPL_THEME1.replace(-999.0, np.nan, inplace=True)
    shapefile.RPL_THEME2.replace(-999.0, np.nan, inplace=True)
    shapefile.RPL_THEME3.replace(-999.0, np.nan, inplace=True)
    shapefile.RPL_THEME4.replace(-999.0, np.nan, inplace=True)

replace_missing(PR_2018_census_tracts_svi_shp)
replace_missing(PR_2018_counties_svi_shp)
replace_missing(PR_2016_census_tracts_svi_shp)
replace_missing(PR_2016_counties_svi_shp)

replace_missing(FL_2018_census_tracts_svi_shp)
replace_missing(FL_2018_counties_svi_shp)
replace_missing(FL_2016_census_tracts_svi_shp)
replace_missing(FL_2016_counties_svi_shp)

replace_missing(TX_2018_census_tracts_svi_shp)
replace_missing(TX_2018_counties_svi_shp)
replace_missing(TX_2016_census_tracts_svi_shp)
replace_missing(TX_2016_counties_svi_shp)
```

```python
# County outlines
PR_county_outline = PR_2018_census_tracts_svi_shp.dissolve(by=PR_2018_census_tracts_svi_shp.columns[4], aggfunc='first')
FL_county_outline = FL_2018_census_tracts_svi_shp.dissolve(by=FL_2018_census_tracts_svi_shp.columns[4], aggfunc='first')
TX_county_outline = TX_2018_census_tracts_svi_shp.dissolve(by=TX_2018_census_tracts_svi_shp.columns[4], aggfunc='first')
```

```python
def plot_svi_map(svi_shp, county_outline, state_name):
    fig, ax = plt.subplots(figsize=(15, 10))

    # Nice legend
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)

    svi_shp.plot(column='RPL_THEMES', ax=ax, legend=True,  cax=cax, cmap='GnBu')
    county_outline.plot(ax=ax, facecolor='none', edgecolor='grey', linewidth=.5)

    # title and layout
    ax.set_title('SVI ' + state_name + '  2018 CDC', fontsize=20)
    ax.axis('off')
    plt.tight_layout()
```

```python
def plot_svi_maps(svi_shp1, svi_shp2, year1, year2, county_outline, state_name):   
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))   

    # Nice legend
    divider1 = make_axes_locatable(ax1)
    cax1 = divider1.append_axes("right", size="5%", pad=0.1)
    svi_shp1.plot(column='RPL_THEMES', ax=ax1, legend=True,  cax=cax1, cmap='GnBu')
    county_outline.plot(ax=ax1, facecolor='none', edgecolor='grey', linewidth=.5)    
    ax1.set_title('SVI ' + state_name + ' ' + year1 + ' CDC', fontsize=20)
    ax1.axis('off')
    
    divider2 = make_axes_locatable(ax2)
    cax2 = divider2.append_axes("right", size="5%", pad=0.1)
    svi_shp2.plot(column='RPL_THEMES', ax=ax2, legend=True,  cax=cax2, cmap='GnBu')
    county_outline.plot(ax=ax2, facecolor='none', edgecolor='grey', linewidth=.5)    
    ax2.set_title('SVI ' + state_name + ' ' + year2 + ' CDC', fontsize=20)
    ax2.axis('off')    
    
    plt.tight_layout()
```

```python
def plot_svi_diff_map(svi_shp1, svi_shp2, county_outline, state_name):
    
    # Merge the shapefiles
    svi_shp2 = svi_shp2.merge(svi_shp1.loc[:, ['FIPS', 'RPL_THEMES']], suffixes=('', '_2016'), on='FIPS')
    svi_shp2['RPL_THEMES_diff'] = svi_shp2['RPL_THEMES'] - svi_shp2['RPL_THEMES_2016']
    svi_shp2['RPL_THEMES_percent_diff'] = (svi_shp2['RPL_THEMES_diff']/svi_shp2['RPL_THEMES']) * 100
    svi_shp2['RPL_THEMES_percent_diff'] = np.where(svi_shp2['RPL_THEMES_percent_diff'] <= 0, np.NAN, svi_shp2['RPL_THEMES_percent_diff'])
    
    fig, ax = plt.subplots(figsize=(15, 10))

    # Nice legend
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    svi_shp2.plot(column='RPL_THEMES_percent_diff', ax=ax, legend=True,  cax=cax, cmap='GnBu', missing_kwds={'color': 'lightgrey'})
    county_outline.plot(ax=ax, facecolor='none', edgecolor='grey', linewidth=.5)

    # title and layout
    ax.set_title(statename + ' SVI Percent Decline from 2016 to 2018', fontsize=20)
    ax.axis('off')
    plt.tight_layout()   
```

```python
plot_svi_maps(PR_2016_counties_svi_shp, PR_2018_counties_svi_shp, '2016', '2018', PR_county_outline, 'Puerto Rico')
```

```python
plot_svi_diff_map(PR_2016_counties_svi_shp, PR_2018_counties_svi_shp, PR_county_outline, 'Puerto Rico')
```

```python
plot_svi_maps(FL_2016_counties_svi_shp, FL_2018_counties_svi_shp, '2016', '2018', FL_county_outline, 'Florida')
```

```python
plot_svi_diff_map(FL_2016_counties_svi_shp, FL_2018_counties_svi_shp, FL_county_outline, 'Florida')
```

```python
plot_svi_maps(TX_2016_counties_svi_shp, TX_2018_counties_svi_shp, '2016', '2018', TX_county_outline, 'Texas')
```

```python
plot_svi_diff_map(TX_2016_counties_svi_shp, TX_2018_counties_svi_shp, TX_county_outline, 'Texas')
```

### DVI

```python
# Load the DVI files
census_tracts_dvi = pd.read_csv('../data/census-tract/census-tract-dvi.csv')
census_tracts_dvi['tractid'] = census_tracts_dvi.censusid.astype(str).str[9:]

PR_census_tracts_dvi = census_tracts_dvi[census_tracts_dvi.tractid.astype(str).str[:2] == '72']
FL_census_tracts_dvi = census_tracts_dvi[census_tracts_dvi.tractid.astype(str).str[:2] == '12']
TX_census_tracts_dvi = census_tracts_dvi[census_tracts_dvi.tractid.astype(str).str[:2] == '48']
```

```python
PR_census_tracts_dvi.describe()
```

```python
# Merge the dvi values with the above shapefiles
PR_census_tracts_svi_shp = PR_2018_census_tracts_svi_shp.merge(
    PR_census_tracts_dvi.loc[:, ['tractid', 'dvi']], how='left', left_on='FIPS', right_on='tractid')
FL_census_tracts_svi_shp = FL_2018_census_tracts_svi_shp.merge(
    FL_census_tracts_dvi.loc[:, ['tractid', 'dvi']], how='left', left_on='FIPS', right_on='tractid')
TX_census_tracts_svi_shp = TX_2018_census_tracts_svi_shp.merge(
    TX_census_tracts_dvi.loc[:, ['tractid', 'dvi']], how='left', left_on='FIPS', right_on='tractid')
```

```python
def plot_svi_dvi_map(census_tracts_svi_shp, county_outline, state_name):   
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))
    
    # Plot SVI
    divider1 = make_axes_locatable(ax1)
    cax1 = divider1.append_axes("right", size="5%", pad=0.1)
    census_tracts_svi_shp.plot(column='RPL_THEMES', ax=ax1, legend=True,  cax=cax1, cmap='GnBu', missing_kwds={'color': 'lightgrey'})
    county_outline.plot(ax=ax1, facecolor='none', edgecolor='grey', linewidth=.5)
    ax1.set_title('SVI ' + state_name + ' 2018', fontsize=20)
    ax1.axis('off')
    
    # Plot DVI
    divider2 = make_axes_locatable(ax2)
    cax2 = divider2.append_axes("right", size="5%", pad=0.1)
    census_tracts_svi_shp.plot(column='dvi', ax=ax2, legend=True,  cax=cax2, cmap='GnBu', missing_kwds={'color': 'lightgrey'})
    county_outline.plot(ax=ax2, facecolor='none', edgecolor='grey', linewidth=.5)
    ax2.set_title('DVI ' + state_name + ' 2017', fontsize=20)
    ax2.axis('off')
    
    plt.tight_layout()
```

```python
plot_svi_dvi_map(PR_census_tracts_svi_shp, PR_county_outline, 'Puerto Rico')
```

```python
plot_svi_dvi_map(FL_census_tracts_svi_shp, FL_county_outline, 'Florida')
```

```python
plot_svi_dvi_map(TX_census_tracts_svi_shp, TX_county_outline, 'Texas')
```
