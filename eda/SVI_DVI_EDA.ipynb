```python
import pandas as pd
import geopandas as gpd
import numpy as np

from matplotlib import pyplot as plt 

# For legend
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

### CDC SVI

```python
# Read the CDC files
PR_census_tracts_svi_shp = gpd.read_file('../data/census-tract/cdc-svi/SVI2018_PuertoRico_tract.shp')
FL_census_tracts_svi_shp = gpd.read_file('../data/census-tract/cdc-svi/SVI2018_FLORIDA_tract.shp')
TX_census_tracts_svi_shp = gpd.read_file('../data/census-tract/cdc-svi/SVI2018_TEXAS_tract.shp')
```

```python
# Convert -999 to nan's
PR_census_tracts_svi_shp.RPL_THEMES.replace(-999.0, np.nan, inplace=True)
PR_census_tracts_svi_shp.RPL_THEME1.replace(-999.0, np.nan, inplace=True)
PR_census_tracts_svi_shp.RPL_THEME2.replace(-999.0, np.nan, inplace=True)
PR_census_tracts_svi_shp.RPL_THEME3.replace(-999.0, np.nan, inplace=True)
PR_census_tracts_svi_shp.RPL_THEME4.replace(-999.0, np.nan, inplace=True)

FL_census_tracts_svi_shp.RPL_THEMES.replace(-999.0, np.nan, inplace=True)
FL_census_tracts_svi_shp.RPL_THEME1.replace(-999.0, np.nan, inplace=True)
FL_census_tracts_svi_shp.RPL_THEME2.replace(-999.0, np.nan, inplace=True)
FL_census_tracts_svi_shp.RPL_THEME3.replace(-999.0, np.nan, inplace=True)
FL_census_tracts_svi_shp.RPL_THEME4.replace(-999.0, np.nan, inplace=True)

TX_census_tracts_svi_shp.RPL_THEMES.replace(-999.0, np.nan, inplace=True)
TX_census_tracts_svi_shp.RPL_THEME1.replace(-999.0, np.nan, inplace=True)
TX_census_tracts_svi_shp.RPL_THEME2.replace(-999.0, np.nan, inplace=True)
TX_census_tracts_svi_shp.RPL_THEME3.replace(-999.0, np.nan, inplace=True)
TX_census_tracts_svi_shp.RPL_THEME4.replace(-999.0, np.nan, inplace=True)
```

```python
# County outlines
PR_county_outline = PR_census_tracts_svi_shp.dissolve(by=PR_census_tracts_svi_shp.columns[4], aggfunc='first')
FL_county_outline = FL_census_tracts_svi_shp.dissolve(by=FL_census_tracts_svi_shp.columns[4], aggfunc='first')
TX_county_outline = TX_census_tracts_svi_shp.dissolve(by=TX_census_tracts_svi_shp.columns[4], aggfunc='first')
```

```python
def plot_map(svi_shp, county_outline, state_name):
    fig, ax = plt.subplots(figsize=(15, 10))

    # Nice legend
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)

    # Plot the demographic and superimpose the county outline
    # Highlight the missing datapoints in grey
    # color scheme: https://matplotlib.org/tutorials/colors/colormaps.html
    svi_shp.plot(column='RPL_THEMES', ax=ax, legend=True,  cax=cax, cmap='GnBu', missing_kwds={'color': 'lightgrey'})
    county_outline.plot(ax=ax, facecolor='none', edgecolor='grey', linewidth=.5)

    # title and layout
    ax.set_title('SVI ' + state_name + '  2018 CDC', fontsize=20)
    ax.axis('off')
    plt.tight_layout()
```

```python
plot_map(PR_census_tracts_svi_shp, PR_county_outline, 'Puerto Rico')
```

```python
plot_map(FL_census_tracts_svi_shp, FL_county_outline, 'Florida')
```

```python
plot_map(TX_census_tracts_svi_shp, TX_county_outline, 'Texas')
```

### DVI

```python
# Load the DVI files
census_tracts_dvi = pd.read_csv('../data/census-tract/census-tract-dvi.csv')
census_tracts_dvi['tractid'] = census_tracts_dvi.censusid.astype(str).str[9:]

PR_census_tracts_dvi = census_tracts_dvi[census_tracts_dvi.tractid.astype(str).str[:2] == '72']
FL_census_tracts_dvi = census_tracts_dvi[census_tracts_dvi.tractid.astype(str).str[:2] == '12']
TX_census_tracts_dvi = census_tracts_dvi[census_tracts_dvi.tractid.astype(str).str[:2] == '48']
```

```python
# Merge the dvi values with the above shapefiles
PR_census_tracts_svi_shp = PR_census_tracts_svi_shp.merge(
    PR_census_tracts_dvi.loc[:, ['tractid', 'dvi']], how='left', left_on='FIPS', right_on='tractid')
FL_census_tracts_svi_shp = FL_census_tracts_svi_shp.merge(
    FL_census_tracts_dvi.loc[:, ['tractid', 'dvi']], how='left', left_on='FIPS', right_on='tractid')
TX_census_tracts_svi_shp = TX_census_tracts_svi_shp.merge(
    TX_census_tracts_dvi.loc[:, ['tractid', 'dvi']], how='left', left_on='FIPS', right_on='tractid')
```

```python
def plot_svi_dvi(census_tracts_svi_shp, county_outline, state_name):   
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))
    
    # Plot SVI
    divider1 = make_axes_locatable(ax1)
    cax1 = divider1.append_axes("right", size="5%", pad=0.1)
    census_tracts_svi_shp.plot(column='RPL_THEMES', ax=ax1, legend=True,  cax=cax1, cmap='GnBu', missing_kwds={'color': 'lightgrey'})
    county_outline.plot(ax=ax1, facecolor='none', edgecolor='grey', linewidth=.5)
    ax1.set_title('SVI ' + state_name + ' 2018', fontsize=20)
    ax1.axis('off')
    
    # Plot DVI
    divider2 = make_axes_locatable(ax2)
    cax2 = divider2.append_axes("right", size="5%", pad=0.1)
    census_tracts_svi_shp.plot(column='dvi', ax=ax2, legend=True,  cax=cax2, cmap='GnBu', missing_kwds={'color': 'lightgrey'})
    county_outline.plot(ax=ax2, facecolor='none', edgecolor='grey', linewidth=.5)
    ax2.set_title('DVI ' + state_name + ' 2017', fontsize=20)
    ax2.axis('off')
    
    plt.tight_layout()
```

```python
plot_svi_dvi(PR_census_tracts_svi_shp, PR_county_outline, 'Puerto Rico')
```

```python
plot_svi_dvi(FL_census_tracts_svi_shp, FL_county_outline, 'Florida')
```

```python
plot_svi_dvi(TX_census_tracts_svi_shp, TX_county_outline, 'Texas')
```
